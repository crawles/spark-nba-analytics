# From NBA Savant

```python
data_dir = 'data/shot_charts'
dfs = []
for yr in range(2011,(2016+1)):
    df = pd.read_csv(os.path.join(data_dir,'shot_chart_{}.csv'.format(yr)))
    df['season_yr'] = yr
    dfs.append(df)
shot_df = pd.concat(dfs)
```

```python
shot_df.columns
```

```python
james_harden = shot_df.query('name == "James Harden" and season_yr == 2013')
james_harden
```

```python
n = 5000
# plt.plot(shot_df.iloc[0:n].x,shot_df.iloc[0:n].y, 'bo', alpha = .05)
sns.jointplot(shot_df.x,shot_df.y,kind = 'hex', stat_func=None, size = 1)
```

```python
# get our colormap for the main kde plot
# Note we can extract a color from cmap to use for 
# the plots that lie on the side and top axes
cmap=plt.cm.YlOrRd_r 

# n_levels sets the number of contour lines for the main kde plot
joint_shot_chart = sns.jointplot(james_harden.x, james_harden.y, stat_func=None,
                                 kind='kde', space=0, color=cmap(0.1),
                                 cmap=cmap, n_levels=50)
```
